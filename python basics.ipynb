```python
from types import WrapperDescriptorType
# FUNCTIONS

def greet(name):
  return f"Hello boy !!!", name
print(greet("Zayed "))

#----------------------------------
def add(z, b=4):
  return z+b
print(add(3))
print(add(3,7))

#---------------------------------

def info(name, age, no):
  print(name, age, no)

info("Zaid", 21, 215478855)
info(age=22, name='zai', no=488885)

#--------------------------------------

sqr = lambda x: x*x
print(sqr(5))
print(sqr(7))

#------------------------------------

n= lambda x: "+ve" if x>0 else "-ve" if x<0 else "zero"
print(n(12))
print(n(-5))

#-------------------------------------

def factorial(n):
  if n==1:
    return 1
  else:
    return n*factorial(n-1)

print(factorial(5))

#-------------------------------------

def decorator(func):
    def wrapp():
        print("before function")
        func()
        print("after function")
    return wrapp

def greet():
    print("Hello, world!")

greet()
```

```python
# Dict Comprehension

keys = ['name', 'age', 'gender']
values = ['Alice', 25, 'F']
mapiing={keys[i] : values[i] for i in range(len(keys))}
print(mapiing)

#----------------------------------------------------------

dicti={'A': 45, 'B': 67, 'C': 89, 'D': 34}
filter= {x:z for x, z in dicti.items() if z>50}
print(filter)

#---------------------------------------------------------

mark={'Alice': 85, 'Bob': 72, 'Charlie': 90}
grade={ name : 'A'
       if marks>80
        else 'B'
        if marks>70
        else 'C' for name , marks in mark.items()}

print(grade)
```
